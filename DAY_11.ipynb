```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('taxis')
df
```

```python
sns.barplot(x='payment',y='total',data=df,color='green',estimator='mean')
plt.show()
plt.tight_layout()
```

```python
sns.histplot(x='total',data=df,color='black')
plt.show()
plt.tight_layout()
```

```python
sns.histplot(x=df['fare'],palette=['red','yellow'],hue=df['payment'])
plt.show()
plt.tight_layout()
```

```python
sns.boxplot(x=df['fare'],color='brown')
plt.show()
plt.tight_layout()
```
